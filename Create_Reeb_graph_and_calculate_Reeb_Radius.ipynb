### Calculate Reeb Radius

```python
import networkx as nx

def reeb_radius_asymmetric_verbose(G):
    """
    Compute and print the asymmetric Reeb radius for all node pairs.

    Parameters:
    G (nx.DiGraph): A directed graph where each node has a `f_vertex` attribute (height).
    """
    print("Asymmetric Reeb Radius:")

    # Iterate over all pairs of nodes
    for start_node in G.nodes:
        for end_node in G.nodes:
            if start_node == end_node:
                # Self-loop case
                print(f"Radius({start_node}, {end_node}) = 0 (self-loop)")
                continue

            try:
                # Find the shortest path
                path = nx.shortest_path(G, source=start_node, target=end_node)
                heights = [G.nodes[node]['f_vertex'] for node in path]
                max_height = max(heights)

                # Calculate radius for both directions
                radius_start_to_end = max_height - G.nodes[start_node]['f_vertex']
                radius_end_to_start = max_height - G.nodes[end_node]['f_vertex']

                # Print the results
                print(f"Radius({start_node}, {end_node}) = {radius_start_to_end}")
                print(f"Radius({end_node}, {start_node}) = {radius_end_to_start}")
            except nx.NetworkXNoPath:
                # No path case
                print(f"Radius({start_node}, {end_node}) = undefined (no path)")
                print(f"Radius({end_node}, {start_node}) = undefined (no path)")
```

```python
import networkx as nx
import heapq

def compute_reeb_radius(G, source):
    """
    Compute the asymmetric Reeb radius from a source node to all other nodes.

    Parameters:
    G (nx.Graph): An undirected graph where each node has a `f_vertex` attribute (height).
    source (str): The source node.

    Returns:
    dict: A dictionary with target nodes as keys and their Reeb radius values as values.
    """
    # Initialize distances to infinity and set the source to 0
    reeb_radius = {node: float('inf') for node in G.nodes}
    reeb_radius[source] = 0

    # Priority queue for Dijkstra-like traversal
    pq = [(0, source)]  # (current_radius, current_node)

    while pq:
        current_radius, current_node = heapq.heappop(pq)

        # Skip if this path is not better
        if current_radius > reeb_radius[current_node]:
            continue

        # Update neighbors
        for neighbor in G.neighbors(current_node):
            max_height = max(G.nodes[current_node]['f_vertex'], G.nodes[neighbor]['f_vertex'])
            new_radius = max(current_radius, abs(max_height - G.nodes[source]['f_vertex']))

            if new_radius < reeb_radius[neighbor]:
                reeb_radius[neighbor] = new_radius
                heapq.heappush(pq, (new_radius, neighbor))

    return reeb_radius

def reeb_radius_asymmetric_verbose(G):
    """
    Compute and print the asymmetric Reeb radius for all node pairs.

    Parameters:
    G (nx.Graph): An undirected graph where each node has a `f_vertex` attribute (height).
    """
    print("Asymmetric Reeb Radius:")

    # Compute Reeb radius for all node pairs
    for source in G.nodes:
        radii = compute_reeb_radius(G, source)
        for target, radius in radii.items():
            if source == target:
                print(f"Radius({source}, {target}) = 0 (self-loop)")
            else:
                print(f"Radius({source}, {target}) = {radius}")

# Create the undirected graph
G = nx.Graph()  # Use nx.Graph() for undirected graphs
G.add_node('g1', f_vertex=24)
G.add_node('g2', f_vertex=20)
G.add_node('g3', f_vertex=15)
G.add_node('g4', f_vertex=12)
G.add_node('g5', f_vertex=8)
G.add_node('g6', f_vertex=6)

G.add_edge('g1', 'g2')
G.add_edge('g2', 'g3')
G.add_edge('g3', 'g4')
G.add_edge('g4', 'g5')
G.add_edge('g4', 'g6')

# Function for Reeb radius
reeb_radius_asymmetric_verbose(G)
```

```python
import networkx as nx

# Create an undirected graph
G = nx.Graph()

# Add nodes with 'f_vertex' attribute
G.add_node('g1', f_vertex=24)
G.add_node('g2', f_vertex=20)
G.add_node('g3', f_vertex=15)
G.add_node('g4', f_vertex=12)
G.add_node('g5', f_vertex=8)
G.add_node('g6', f_vertex=6)

# Add edges
G.add_edge('g1', 'g2')
G.add_edge('g2', 'g3')
G.add_edge('g3', 'g4')
G.add_edge('g4', 'g5')
G.add_edge('g4', 'g6')
label_to_int = {label: idx for idx, label in enumerate(G.nodes())}
```

```python
import gudhi

# Initialize a SimplexTree
st = gudhi.SimplexTree()

# Add edges to the simplex tree
for edge in G.edges():
    simplex = [label_to_int[edge[0]], label_to_int[edge[1]]]
    st.insert(simplex)

# Add nodes to the simplex tree
for node in G.nodes():
    simplex = [label_to_int[node]]
    st.insert(simplex)

```

```python
# Compute persistence
st.compute_persistence()

# Retrieve the persistence intervals
persistence_intervals = st.persistence()
```

```python
import numpy as np

# Define the distance matrices
D_R_f = np.array([
    [0, 3, 2, 6, 4.5, 9, 8, 11],
    [3, 0, 1, 3, 1.5, 6, 5, 8],
    [2, 1, 0, 4, 2.5, 7, 6, 9],
    [6, 3, 4, 0, 4.5, 6, 4, 5],
    [4.5, 1.5, 2.5, 4.5, 0, 7.5, 5.5, 6.5],
    [9, 6, 7, 6, 7.5, 0, 5, 8],
    [8, 5, 6, 4, 5.5, 5, 0, 6],
    [11, 8, 9, 5, 6.5, 8, 6, 0]
])

D_R_g = np.array([
    [0, 2, 4.5, 6, 8, 9],
    [2, 0, 2.5, 4, 6, 7],
    [4.5, 2.5, 0, 1.5, 3.5, 4.5],
    [6, 4, 1.5, 0, 2, 3],
    [8, 6, 3.5, 2, 0, 5],
    [9, 7, 4.5, 3, 5, 0]
])

# Define the optimal transport matrix
T = np.array([
    [0.2558, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000],
    [0.0194, 0.0926, 0.0043, 0.0000, 0.0000, 0.0000],
    [0.05815, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000],
    [0.0000, 0.0000, 0.03015, 0.0396, 0.0000, 0.0000],
    [0.0000, 0.0000, 0.05815, 0.0000, 0.0000, 0.0000],
    [0.0000, 0.0000, 0.0000, 0.0000, 0.06975, 0.0000],
    [0.0000, 0.0000, 0.0000, 0.03445, 0.0043, 0.07755],
    [0.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.2558]
])

# Initialize the squared GW distance
gw_distance_squared = 0.0

# Iterate over all indices
for i in range(D_R_f.shape[0]):
    for j in range(D_R_f.shape[0]):
        for k in range(D_R_g.shape[0]):
            for l in range(D_R_g.shape[0]):
                # Compute the squared loss
                loss = (D_R_f[i, j] - D_R_g[k, l]) ** 2
                # Accumulate the weighted loss
                gw_distance_squared += loss * T[i, k] * T[j, l]

# The GW distance is the square root of the accumulated sum
gw_distance = np.sqrt(gw_distance_squared)

print(f"Gromov-Wasserstein distance: {gw_distance}")
```

```python
import pandas as pd
import networkx as nx
import heapq

# Define the graph with correct node attributes
F = nx.Graph()
F.add_node('f1', f_vertex=24)
F.add_node('f2', f_vertex=18)
F.add_node('f3', f_vertex=20)
F.add_node('f4', f_vertex=12)
F.add_node('f5', f_vertex=15)
F.add_node('f6', f_vertex=6)
F.add_node('f7', f_vertex=8)
F.add_node('f8', f_vertex=2)

F.add_edge('f1', 'f2')
F.add_edge('f2', 'f4')
F.add_edge('f2', 'f5')
F.add_edge('f3', 'f5')
F.add_edge('f5', 'f7')
F.add_edge('f4', 'f6')
F.add_edge('f4', 'f7')
F.add_edge('f7', 'f8')

# Correcting the handling of direct neighbors to align with the definition
def reeb_radius_correct(F, root):
    # Initialize Reeb radius dictionary
    rho_g = {v: float('inf') for v in F.nodes}
    rho_g[root] = 0

    # Priority queue for Dijkstra-style processing
    Q = [(0, root)]  # (Reeb radius, node)

    while Q:
        current_radius, v = heapq.heappop(Q)
        for neighbor in F.neighbors(v):
            # Compute the updated radius for the neighbor
            d_m = abs(F.nodes[root]['f_vertex'] - F.nodes[neighbor]['f_vertex'])
            updated_radius = max(rho_g[v], d_m)
            if updated_radius < rho_g[neighbor]:
                rho_g[neighbor] = updated_radius
                heapq.heappush(Q, (updated_radius, neighbor))

    return rho_g

# Recomputing the matrix with the corrected logic
def compute_corrected_reeb_radius_matrix(F):
    nodes = list(F.nodes)
    matrix = pd.DataFrame(float('inf'), index=nodes, columns=nodes)
    for node in nodes:
        radii = reeb_radius_correct(F, node)
        for other_node, radius in radii.items():
            matrix.loc[node, other_node] = radius
    return matrix

# Verify the graph's node attributes
print("Graph Nodes with Attributes:")
print(F.nodes(data=True))

# Compute the corrected Reeb radius matrix
reeb_radius_matrix_corrected = compute_corrected_reeb_radius_matrix(F)
print("\nCorrected Reeb Radius Matrix:")
print(reeb_radius_matrix_corrected)
```

```python
# Final adjustment to ensure correctness of Reeb radius logic
def reeb_radius_proper_final(F, root):
    # Initialize Reeb radius dictionary
    rho_g = {v: float('inf') for v in F.nodes}
    rho_g[root] = 0

    # Priority queue for Dijkstra-style processing
    Q = [(0, root)]  # (Reeb radius, node)

    while Q:
        current_radius, v = heapq.heappop(Q)
        for neighbor in F.neighbors(v):
            # Compute the updated radius for the neighbor
            d_m = abs(F.nodes[root]['f_vertex'] - F.nodes[neighbor]['f_vertex'])
            updated_radius = max(rho_g[v], d_m)

            # Ensure all paths from the root correctly handle the supremum logic
            if F.nodes[root]['f_vertex'] >= max(F.nodes[v]['f_vertex'], F.nodes[neighbor]['f_vertex']):
                updated_radius = 0

            if updated_radius < rho_g[neighbor]:
                rho_g[neighbor] = updated_radius
                heapq.heappush(Q, (updated_radius, neighbor))

    return rho_g

# Recomputing the matrix with the proper logic for paths from the highest node
def compute_final_proper_reeb_radius_matrix(F):
    nodes = list(F.nodes)
    matrix = pd.DataFrame(float('inf'), index=nodes, columns=nodes)
    for node in nodes:
        radii = reeb_radius_proper_final(F, node)
        for other_node, radius in radii.items():
            matrix.loc[node, other_node] = radius
    return matrix

# Compute and display the final corrected Reeb radius matrix
reeb_radius_matrix_proper_final = compute_final_proper_reeb_radius_matrix(F)
reeb_radius_matrix_proper_final
```

```python
import networkx as nx
import gudhi as gd

# Step 1: Define the graph F with numerical node names
F = nx.Graph()

F.add_node(0, f_vertex=24)
F.add_node(1, f_vertex=18)
F.add_node(2, f_vertex=20)
F.add_node(3, f_vertex=12)
F.add_node(4, f_vertex=15)
F.add_node(5, f_vertex=6)
F.add_node(6, f_vertex=8)
F.add_node(7, f_vertex=2)

F.add_edge(0, 1)
F.add_edge(1, 3)
F.add_edge(1, 4)
F.add_edge(2, 4)
F.add_edge(4, 6)
F.add_edge(3, 5)
F.add_edge(3, 6)
F.add_edge(6, 7)

# Step 2: Build the SimplexTree
st = gd.SimplexTree()

# Add nodes with filtration values
for node, attr in F.nodes(data=True):
    st.insert([node], filtration=attr['f_vertex'])

# Add edges with filtration values
for u, v in F.edges:
    filtration_value = max(F.nodes[u]['f_vertex'], F.nodes[v]['f_vertex'])
    st.insert([u, v], filtration=filtration_value)

# Step 3: Make filtration non-decreasing
st.make_filtration_non_decreasing()

# Step 4: Compute persistence diagrams
persistence = st.persistence()

# Step 5: Automatically identify node pairs for birth and death
node_pairs_auto = {}
for idx, (dim, (birth, death)) in enumerate(persistence):
    node_pairs_auto[f"Feature {idx} (H_{dim})"] = {"birth": [], "death": []}
    for simplex, filtration in st.get_filtration():
        if filtration == birth:
            node_pairs_auto[f"Feature {idx} (H_{dim})"]["birth"].append(simplex)
        if filtration == death:
            node_pairs_auto[f"Feature {idx} (H_{dim})"]["death"].append(simplex)

# Display the node pairs
node_pairs_auto
```

```python
# Extract and explicitly pair nodes for each feature
node_pairings = {}
for idx, (dim, (birth, death)) in enumerate(persistence):
    feature_label = f"Feature {idx} (H_{dim})"

    # Get simplices responsible for birth and death
    birth_simplices = [
        simplex for simplex, filtration in st.get_filtration() if filtration == birth
    ]
    death_simplices = [
        simplex for simplex, filtration in st.get_filtration() if filtration == death
    ]

    # Pair nodes or edges based on birth and death simplices
    if birth_simplices and death_simplices:
        birth_pair = tuple(sorted(birth_simplices[0]))  # Sort for consistent output
        death_pair = tuple(sorted(death_simplices[0]))  # Sort for consistent output
    elif birth_simplices:
        birth_pair = tuple(sorted(birth_simplices[0]))
        death_pair = None
    else:
        birth_pair = None
        death_pair = None

    # Store results
    node_pairings[feature_label] = {"birth_pair": birth_pair, "death_pair": death_pair}

# Display the explicitly paired nodes
node_pairings
```

```python
import networkx as nx
import gudhi as gd

# Step 1: Define the graph F with numerical node names
F = nx.Graph()

F.add_node(0, f_vertex=24)
F.add_node(1, f_vertex=18)
F.add_node(2, f_vertex=20)
F.add_node(3, f_vertex=12)
F.add_node(4, f_vertex=15)
F.add_node(5, f_vertex=6)
F.add_node(6, f_vertex=8)
F.add_node(7, f_vertex=2)

F.add_edge(0, 1)
F.add_edge(1, 3)
F.add_edge(1, 4)
F.add_edge(2, 4)
F.add_edge(4, 6)
F.add_edge(3, 5)
F.add_edge(3, 6)
F.add_edge(6, 7)

# Step 2: Create a mapping between node names and integers
node_mapping = {node: idx for idx, node in enumerate(F.nodes)}
mapped_graph = nx.relabel_nodes(F, node_mapping)

# Step 3: Build the SimplexTree
st = gd.SimplexTree()

# Add nodes with filtration values
for node, attr in mapped_graph.nodes(data=True):
    st.insert([node], filtration=attr['f_vertex'])

# Add edges with filtration values
for u, v in mapped_graph.edges:
    filtration_value = max(mapped_graph.nodes[u]['f_vertex'], mapped_graph.nodes[v]['f_vertex'])
    st.insert([u, v], filtration=filtration_value)

# Step 4: Make filtration non-decreasing
st.make_filtration_non_decreasing()

# Step 5: Compute extended persistence
st.extend_filtration()
extended_persistence = st.extended_persistence()

# Correctly mapping node filtration values
node_filtrations = {node: F.nodes[node]['f_vertex'] for node in F.nodes}
inverse_node_mapping = {v: k for k, v in node_mapping.items()}  # Map integers back to original node names

# Function to find the closest matching nodes for scalar or tuple values
def find_closest_mapped_nodes(value, tolerance=1e-6):
    if isinstance(value, tuple):
        # For tuple values, compare both components
        matches = [
            inverse_node_mapping[node]
            for node, filtration in node_filtrations.items()
            if any(abs(filtration - v) < tolerance for v in value)
        ]
    else:
        # For scalar values, compare directly
        matches = [
            inverse_node_mapping[node]
            for node, filtration in node_filtrations.items()
            if abs(filtration - value) < tolerance
        ]
    print(f"Matching nodes for value {value}: {matches}")  # Debugging
    return matches


# Map persistence diagrams to node pairs
mapped_persistence_pairs = {}
for diagram_type, dgm in zip(["Ordinary", "Relative", "Extended+", "Extended-"], extended_persistence):
    print(f"\nProcessing Diagram: {diagram_type}")  # Debugging
    mapped_persistence_pairs[diagram_type] = []
    for birth, death in dgm:
        print(f"  Birth: {birth}, Death: {death}")  # Debugging
        birth_nodes = find_closest_mapped_nodes(birth)
        death_nodes = find_closest_mapped_nodes(death)
        if birth_nodes and death_nodes:
            mapped_persistence_pairs[diagram_type].append((birth_nodes[0], death_nodes[0]))
        elif birth_nodes:
            mapped_persistence_pairs[diagram_type].append((birth_nodes[0], None))
        elif death_nodes:
            mapped_persistence_pairs[diagram_type].append((None, death_nodes[0]))

# Display results
print("\nMapped Persistence Pairs:")
print(mapped_persistence_pairs)
```
