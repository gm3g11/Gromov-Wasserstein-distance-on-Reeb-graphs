Three steps to calculate Gromov-Wasserstein disntance between $R_f$ and $R_g$:

Step 1: Compute the Smoothed Symmetric Reeb Radius.

##Step 1: Compute the Smoothed Symmetric Reeb Radius.

```python
## The Reeb Radius algorithm is based on the algorithm 1 from
## Justin's paper: Stability and Approximations for Decorated Reeb Spaces

import pandas as pd
import networkx as nx
import heapq


def reeb_radius(F, root):
    # Initialize Reeb radius dictionary
    rho_g = {v: float('inf') for v in F.nodes}
    rho_g[root] = 0

    # Priority queue for Dijkstra-style processing
    Q = [(0, root)]  # (Reeb radius, node)

    while Q:
        current_radius, v = heapq.heappop(Q)
        for neighbor in F.neighbors(v):
            # Compute the updated radius for the neighbor
            d_m = abs(F.nodes[root]['f_vertex'] - F.nodes[neighbor]['f_vertex'])
            updated_radius = max(rho_g[v], d_m)

            # Ensure all paths from the root correctly handle the supremum logic
            if F.nodes[root]['f_vertex'] >= max(F.nodes[v]['f_vertex'], F.nodes[neighbor]['f_vertex']):
                updated_radius = 0

            if updated_radius < rho_g[neighbor]:
                rho_g[neighbor] = updated_radius
                heapq.heappush(Q, (updated_radius, neighbor))

    return rho_g

# Recomputing the matrix with the proper logic for paths from the highest node
def compute_reeb_radius_matrix(F):
    nodes = list(F.nodes)
    matrix = pd.DataFrame(float('inf'), index=nodes, columns=nodes)
    for node in nodes:
        radii = reeb_radius(F, node)
        for other_node, radius in radii.items():
            matrix.loc[node, other_node] = radius
    return matrix

```

## Step 2: Compute the Persistence Diagram-Based Borel Probability

```python
import networkx as nx
import gudhi as gd

def persistent_probability(graph):


    # Step 1: Create a mapping between node names and integers
    node_mapping = {node: idx for idx, node in enumerate(graph.nodes)}
    mapped_graph = nx.relabel_nodes(graph, node_mapping)

    # Step 2: Build the SimplexTree
    st = gd.SimplexTree()

    # Add nodes with filtration values
    for node, attr in mapped_graph.nodes(data=True):
        st.insert([node], filtration=attr['f_vertex'])

    # Add edges with filtration values
    for u, v in mapped_graph.edges:
        filtration_value = max(mapped_graph.nodes[u]['f_vertex'], mapped_graph.nodes[v]['f_vertex'])
        st.insert([u, v], filtration=filtration_value)

    # Step 3: Make filtration non-decreasing and compute extended persistence
    st.make_filtration_non_decreasing()
    st.extend_filtration()
    extended_persistence = st.extended_persistence()
    print("Extended persistence:", extended_persistence)
    # Step 4: Map persistence diagrams to node pairs
    node_filtrations = {node: graph.nodes[node]['f_vertex'] for node in graph.nodes}
    inverse_node_mapping = {v: k for k, v in node_mapping.items()}  # Map integers back to original node names

    mapped_persistence_pairs = {}
    diagram_types = ["Ordinary", "Relative", "Extended+", "Extended-"]
    for diagram_type, dgm in zip(diagram_types, extended_persistence):
        mapped_persistence_pairs[diagram_type] = []
        for i in range(len(dgm)):
            birth = dgm[i][1][0]  # Extract birth value
            death = dgm[i][1][1]  # Extract death value

            birth_nodes = [
                inverse_node_mapping[node]
                for node, filtration in node_filtrations.items()
                if abs(filtration - birth) < 1e-6
            ]
            death_nodes = [
                inverse_node_mapping[node]
                for node, filtration in node_filtrations.items()
                if abs(filtration - death) < 1e-6
            ]

            if birth_nodes and death_nodes:
                mapped_persistence_pairs[diagram_type].append((birth_nodes[0], death_nodes[0]))
            elif birth_nodes:
                mapped_persistence_pairs[diagram_type].append((birth_nodes[0], None))
            elif death_nodes:
                mapped_persistence_pairs[diagram_type].append((None, death_nodes[0]))

    # Step 5: Calculate lifespan and normalize
    lifespans = {}
    for diagram_type, pairs in mapped_persistence_pairs.items():
        for birth, death in pairs:
            if birth is not None and death is not None:
                # Calculate lifespan as the difference between death and birth filtration values
                birth_filtration = node_filtrations[birth]
                death_filtration = node_filtrations[death]
                lifespan = abs(death_filtration - birth_filtration)
                lifespans[(birth, death)] = lifespan

    # Calculate the total sum of all lifespans
    total_lifespan = sum(lifespans.values())

    # Normalize lifespans using the total sum
    normalized_lifespans_sum = {
        pair: lifespan / total_lifespan if total_lifespan != 0 else 0
        for pair, lifespan in lifespans.items()
    }

    # Step 6: Recalculate node weights
    node_weights = {}
    if normalized_lifespans_sum:
        for (node1, node2), normalized_lifespan in normalized_lifespans_sum.items():
            # Divide the lifespan equally between the two nodes
            weight = normalized_lifespan / 2
            node_weights[node1] = node_weights.get(node1, 0) + weight
            node_weights[node2] = node_weights.get(node2, 0) + weight

    return mapped_persistence_pairs, node_weights


```

## Step 3: Identify the Optimal Transport Plan and calculate $GW_p$

```python
import numpy as np
import ot
import matplotlib.pyplot as plt
import seaborn as sns

def convert_weights_to_vector(weights_dict):
    """
    Converts a dictionary of node weights into a vector sorted by node IDs.

    Args:
        weights_dict (dict): A dictionary where keys are node IDs and values are weights.

    Returns:
        np.ndarray: A vector with weights arranged in the order of node IDs.
    """
    return np.array([weights_dict[i] for i in sorted(weights_dict.keys())])




# Define the distance matrices for R_f and R_g
D_R_f= res_rf.to_numpy()/2
D_R_g= res_rg.to_numpy()/2
# Define the probability distributions for R_f and R_g
P_R_f= convert_weights_to_vector(F_node_weights)
P_R_g= convert_weights_to_vector(G_node_weights)
# Compute the Gromov-Wasserstein distance
gw_distance, log = ot.gromov.gromov_wasserstein2(
    D_R_f, D_R_g, P_R_f, P_R_g, 'square_loss', log=True
)

# Extract the optimal transport matrix from the log
transport_matrix = log['T']

gw_distance = np.sqrt(gw_distance)

print(f'Gromov-Wasserstein distance: {gw_distance}')
print(transport_matrix)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Function to draw Reeb graphs and transport connections
def plot_reeb_graphs_with_connections(F, G, transport_matrix, labels_R_f, labels_R_g):
    fig, ax = plt.subplots(figsize=(15, 8))

    # Draw the Reeb graph F
    pos_F = nx.spring_layout(F, seed=42)  # Layout for F
    nx.draw(
        F,
        pos_F,
        with_labels=True,
        labels={i: labels_R_f[i] for i in F.nodes},
        node_color='lightblue',
        edge_color='black',
        node_size=500,
        font_size=10,
        ax=ax,
    )

    # Offset for G to place it side by side
    pos_G = {node: (x + 2.5, y) for node, (x, y) in nx.spring_layout(G, seed=42).items()}
    nx.draw(
        G,
        pos_G,
        with_labels=True,
        labels={i: labels_R_g[i] for i in G.nodes},
        node_color='lightgreen',
        edge_color='black',
        node_size=500,
        font_size=10,
        ax=ax,
    )

    # Draw transport connections
    for i, f_node in enumerate(F.nodes):
        for j, g_node in enumerate(G.nodes):
            weight = transport_matrix[i, j]
            if weight > 0.01:  # Only draw connections with significant weight
                x = [pos_F[f_node][0], pos_G[g_node][0]]
                y = [pos_F[f_node][1], pos_G[g_node][1]]
                ax.plot(x, y, linestyle='--', linewidth=weight * 5, color='orange', alpha=0.7)

    # Add legend and title
    ax.set_title('Reeb Graphs with Optimal Transport Connections', fontsize=16)
    plt.show()


# Generate the labels
labels_R_f = [f'f{i+1}' for i in range(len(F.nodes))]
labels_R_g = [f'g{i+1}' for i in range(len(G.nodes))]

# Call the function to plot
plot_reeb_graphs_with_connections(F, G, transport_matrix, labels_R_f, labels_R_g)
```

```python
def plot_reeb_graphs_with_connections_fixed(F, G, transport_matrix, labels_R_f, labels_R_g):
    """
    Plot the Reeb graphs F and G using cereeberus and overlay the optimal transport connections in a single image.
    """
    fig, ax = plt.subplots(figsize=(15, 10))

    # Manually set the axis for a single figure
    plt.axis('off')

    # Draw Reeb graph F on the same axes
    F.draw(cpx=8, with_labels=False, ax=ax)
    for node, (x, y) in F.pos_f.items():
        ax.text(
            x + 0.02, y, node,
            fontsize=14,
            ha='left',
            va='center'
        )

    # Draw Reeb graph G on the same axes
    G.draw(cpx=1, with_labels=False, ax=ax)
    for node, (x, y) in G.pos_f.items():
        ax.text(
            x + 0.02, y, node,
            fontsize=14,
            ha='left',
            va='center'
        )

    # Overlay transport connections
    for i, f_node in enumerate(F.nodes):
        for j, g_node in enumerate(G.nodes):
            weight = transport_matrix[i, j]
            if weight > 0.01:  # Only plot significant weights
                x = [F.pos_f[f_node][0], G.pos_f[g_node][0]]
                y = [F.pos_f[f_node][1], G.pos_f[g_node][1]]
                ax.plot(x, y, linestyle='--', linewidth=weight * 5, color='orange', alpha=0.7)

    # Set the title for the single image
    ax.set_title('Reeb Graphs with Optimal Transport Connections', fontsize=18)

    # Save the combined plot
    plt.savefig('./reeb_graphs_combined.jpg', bbox_inches='tight')
    plt.show()

# Call the function
plot_reeb_graphs_with_connections_fixed(F, G, transport_matrix, labels_R_f, labels_R_g)
```
